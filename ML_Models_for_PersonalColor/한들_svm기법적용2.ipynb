```python
import pandas as pd
import numpy as np

air = pd.read_csv("./군집화_y값완성_with일시.csv", header=0)
air.head()
```

```python
# 열 삭제하기 위해 drop 사용
air = air.drop(['Unnamed: 0'], axis = 1) # 삭제한 새로운 데이터프레임을 새로운 변수에 저장
air.head()
```

```python
# Ver1 : 대기 구성으로만 설정한 것

# 분류 분석에 사용할 변수 선택
X_ver1 = nair[['SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25']] # 독립변수
Y_ver1 = nair['대기질 등급'] # 종속변수

# KNN 분류 분석을 수행하기 위해 설명변수를 정규화(평균0, 표준편차1를 갖는 정규화 분포로)
from sklearn import preprocessing
X_ver1 = preprocessing.StandardScaler().fit(X_ver1).transform(X_ver1)

# Data set을 trainning과 test로 7:3으로 분리
from sklearn.model_selection import train_test_split
X_ver1_train, X_ver1_test, Y_ver1_train, Y_ver1_test = train_test_split(X_ver1, Y_ver1, test_size=0.3, random_state=10)
```

```python
from sklearn import svm
svm_model = svm.SVC(kernel = 'rbf')
svm_model.fit(X_ver1_train, Y_ver1_train)  #학습, 모델 생성

Y_ver1_predict = svm_model.predict(X_ver1_test)
print(Y_ver1_predict[0:10])
print(Y_ver1_test.values[0:10])

from sklearn import metrics
svm_matrix = metrics.confusion_matrix(Y_ver1_test, Y_ver1_predict)
print(svm_matrix)

svm_report = metrics.classification_report(Y_ver1_test, Y_ver1_predict)
print(svm_report)
```
