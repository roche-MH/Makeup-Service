```python
import pandas as pd
import numpy as np

air = pd.read_csv("./군집화_y값완성_with일시.csv", header=0)
print(air.head())

# 열 삭제하기 위해 drop 사용
air = air.drop(['Unnamed: 0'], axis = 1) # 삭제한 새로운 데이터프레임을 새로운 변수에 저장
```

```python
# Ver1 : 대기 구성으로만 설정한 것

# 분류 분석에 사용할 변수 선택
X_ver1 = air[['SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25']] # 독립변수
Y_ver1 = air['대기질 등급'] # 종속변수

# KNN 분류 분석을 수행하기 위해 설명변수를 정규화(평균0, 표준편차1를 갖는 정규화 분포로)
from sklearn import preprocessing
X_ver1 = preprocessing.StandardScaler().fit(X_ver1).transform(X_ver1)

# Data set을 trainning과 test로 7:3으로 분리
from sklearn.model_selection import train_test_split
X_ver1_train, X_ver1_test, Y_ver1_train, Y_ver1_test = train_test_split(X_ver1, Y_ver1, test_size=0.3, random_state=10)

# KNN 분류 분석으로 모델 생성
from sklearn.neighbors import KNeighborsClassifier
knn_ver1 = KNeighborsClassifier(n_neighbors=9) # K의 개수를 지정해야 함
knn_ver1.fit(X_ver1_train, Y_ver1_train) # 학습시키기 위해 트레이닝 데이터 입력

# 학습 데이터로부터 생성된 모델의 예측값 생성
Y_ver1_predict = knn_ver1.predict(X_ver1_test)

from sklearn import metrics
knn_ver1_matrix = metrics.confusion_matrix(Y_ver1_test, Y_ver1_predict)
print(knn_ver1_matrix) # 결과값 행렬로 출력
knn_ver1_report = metrics.classification_report(Y_ver1_test, Y_ver1_predict)
print(knn_ver1_report)
```

```python
# Ver2 : 대기 구성으로만 설정한 것

X_ver2 = air[['기온(°C)', '풍속(m/s)', '풍향(16방위)', '습도(%)']] # 독립변수
Y_ver2 = air['대기질 등급'] # 종속변수

# KNN 분류 분석을 수행하기 위해 설명변수를 정규화(평균0, 표준편차1를 갖는 정규화 분포로)
from sklearn import preprocessing
X_ver2 = preprocessing.StandardScaler().fit(X_ver2).transform(X_ver2)

# Data set을 trainning과 test로 7:3으로 분리
from sklearn.model_selection import train_test_split
X_ver2_train, X_ver2_test, Y_ver2_train, Y_ver2_test = train_test_split(X_ver2, Y_ver2, test_size=0.3, random_state=10)

# KNN 분류 분석으로 모델 생성
from sklearn.neighbors import KNeighborsClassifier
knn_ver2 = KNeighborsClassifier(n_neighbors=9) # K의 개수를 지정해야 함
knn_ver2.fit(X_ver2_train, Y_ver2_train) # 학습시키기 위해 트레이닝 데이터 입력

# 학습 데이터로부터 생성된 모델의 예측값 생성
Y_ver2_predict = knn_ver2.predict(X_ver2_test)

from sklearn import metrics
knn_ver2_matrix = metrics.confusion_matrix(Y_ver2_test, Y_ver2_predict)
print(knn_ver2_matrix) # 결과값 행렬로 출력
knn_ver2_report = metrics.classification_report(Y_ver2_test, Y_ver2_predict)
print(knn_ver2_report)
```

```python
# Ver3 : 대기 구성 + 기온&바람&습도로 설정한 것

X_ver3 = air[['SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25', '기온(°C)', '풍속(m/s)', '풍향(16방위)', '습도(%)']] # 독립변수
Y_ver3 = air['대기질 등급'] # 종속변수

# KNN 분류 분석을 수행하기 위해 설명변수를 정규화(평균0, 표준편차1를 갖는 정규화 분포로)
from sklearn import preprocessing
X_ver3 = preprocessing.StandardScaler().fit(X_ver3).transform(X_ver3)

# Data set을 trainning과 test로 7:3으로 분리
from sklearn.model_selection import train_test_split
X_ver3_train, X_ver3_test, Y_ver3_train, Y_ver3_test = train_test_split(X_ver3, Y_ver3, test_size=0.3, random_state=10)

# KNN 분류 분석으로 모델 생성
from sklearn.neighbors import KNeighborsClassifier
knn_ver3 = KNeighborsClassifier(n_neighbors=9) # K의 개수를 지정해야 함
knn_ver3.fit(X_ver3_train, Y_ver3_train) # 학습시키기 위해 트레이닝 데이터 입력

# 학습 데이터로부터 생성된 모델의 예측값 생성
Y_ver3_predict = knn_ver3.predict(X_ver3_test)

from sklearn import metrics
knn_ver3_matrix = metrics.confusion_matrix(Y_ver3_test, Y_ver3_predict)
print(knn_ver3_matrix) # 결과값 행렬로 출력
knn_ver3_report = metrics.classification_report(Y_ver3_test, Y_ver3_predict)
print(knn_ver3_report)
```
